```python
# Define the forward pass to make a reasonable use of the attributes
# you defined before. Follow the same reasoning we used in class
class NeuralNetwork(nn.Module):
    def __init__(self,input_dim, num_hidden1, num_hidden2):
        super().__init__()
        self.fc1 = nn.Linear(input_dim, num_hidden1)
        self.fc2 = nn.Linear(num_hidden1, num_hidden2)
        self.fc3 = nn.Linear(num_hidden2, 1)
        self.sigmoid = nn.Sigmoid()

    def forward(self,x):
        output1 = self.fc1(x)
        act1 = self.sigmoid(output1)
        output2 = self.fc2(act1)
        act2 = self.sigmoid(output2)
        output3 = self.fc3(act2)
        final_output = self.sigmoid(output3)
        return final_output  
```

```python
# For training a model, use the following optimizer and loss
model = NeuralNetwork(2, 16, 8)
optimizer = torch.optim.Adam(model.parameters(), lr=0.01)
loss = nn.BCELoss()
```

```python
df=pd.read_csv('./data.csv',header=None)
```

```python
X=df.iloc[0:,0:-1].values
y=df.iloc[0:,-1].values
print(y)
```

```python
input_tensor = torch.from_numpy(X).float()
labels = torch.from_numpy(y).float()
labels = labels.view(-1, 1)
print(labels.shape)
```

```python
# Steps in the training loop

# reset optimazer
optimizer.zero_grad()

# feed model
pred = model.forward(input_tensor)

# compute the loss
loss_value = loss(pred, labels)

# backpropagation
loss_value.backward()

# save the updated weights
optimizer.step()

print(f"loss value: {loss_value:.2}")
```

```python
epochs = 10000

for i in range(epochs):
    
    #reset optimazer
    optimizer.zero_grad()

    # feed model
    pred = model.forward(input_tensor)

    # compute the loss
    loss_value = loss(pred, labels)

    # backpropagation
    loss_value.backward()

    # save the updated weights
    optimizer.step()

    print(f"loss value: {loss_value:.4}")
```
